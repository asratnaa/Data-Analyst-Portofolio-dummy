```python
import pandas as pd
import random

# Buat data dummy manual
data = {
    "Transaction_ID": [f"T{i:03}" for i in range(1, 31)],  # ID transaksi berurutan
    "Customer_Name": [f"Customer {i}" for i in range(1, 31)],  # Nama pelanggan sederhana
    "Product": random.choices(["Laptop", "Smartphone", "Headphones", "Keyboard"], k=30),
    "Price": random.choices([1000000, 2000000, 3000000, 5000000], k=30),
    "Quantity": [random.randint(1, 5) for _ in range(30)],  # Kuantitas
    "Date": pd.date_range(start="2023-01-01", periods=30).tolist(),  # Tanggal berurutan
    "Region": random.choices(["North", "South", "East", "West"], k=30),  # Wilayah
}

# DataFrame dari dictionary
df = pd.DataFrame(data)
```

```python
data_dummy = df.copy() 
data_dummy.head()
```

```python
data_dummy["Total"] = data_dummy["Quantity"] * data_dummy["Price"]
```

```python
sales = data_dummy.groupby(["Product","Region"])
sales_total = sales.agg({"Quantity" : "sum","Total": "sum"})
sales_total
```

```python
def produk (x):
    if x < 2_000_000:
        return "Low"
    elif x < 4_000_000 :
        return "Medium"
    else:
        return "High"

data_dummy["Kategori"] = data_dummy["Price"].apply(produk)
```

```python
data_dummy.to_csv("Data dummy",index=False)
```
